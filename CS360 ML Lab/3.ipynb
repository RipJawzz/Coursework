```python
import pandas as pd
import numpy as np
df= pd.read_csv('/content/drive/MyDrive/Other-Material/A3_Tuesday.csv',header=0).iloc[:,1:]
mat=df.to_numpy()
print("Shape of matrix loaded:",mat.shape)
print("Number of Features:",mat.shape[1]-1)
print("Number of Patterns:",mat.shape[0])
op=mat[:,-1]
print("Range:",np.min(op),np.max(op))
```

2.

```python
ten_pc=int(mat.shape[0]/10)
t_mat=mat.copy()
np.random.shuffle(t_mat) 
splits=np.arange(ten_pc,mat.shape[0]-ten_pc+1,ten_pc)
print("dataset split in X=10:10:90 and Y=90:-10:10")
for i in splits:
  X=t_mat[:i,:]
  Y=t_mat[i:,:]
  print("X:",X.shape,"Y:",Y.shape)
```

3.

```python
from scipy import stats

print("Means:")
print("Row wise:",mat.mean(axis=1))
print("Column wise:",mat.mean(axis=0))
print("Overall:",mat.reshape(1,mat.shape[1]*mat.shape[0]).mean(axis=1))

print("\nMedian")
print("Row Wise",np.median(mat,axis=1))
print("Column Wise",np.median(mat,axis=0))
print("Overall",np.median(mat.reshape(1,mat.shape[1]*mat.shape[0]),axis=1))

print("\nMode")
print("Row Wise:",stats.mode(np.transpose(mat))[0])
print("Column Wise:",stats.mode(mat)[0])
print("Overall:",stats.mode(mat.reshape(mat.shape[1]*mat.shape[0],1))[0])

print("\nStandard Deviation")
print("Row Wise:",np.std(mat,axis=1))
print("Column Wise:",np.std(mat,axis=0))
print("Overall:",np.std(mat))
```
