```python
import pandas as pd
import numpy as np
import random
```

```python
def perceptron(w,X):
    res = X.dot(w)
    if res>=0:
        return 1
    else:
        return 0

def check_convergence(w,P,N):
    for i in P:
        if perceptron(w,i)==0:
            return False
    for i in N:
        if perceptron(w,i)==1:
            return False
    return True
```

```python
w=np.random.randint(0,1,3).reshape(3,1)
P=[]
P.append(np.array([1,1,1]))
N=[]
N.append(np.array([1,1,0]))
N.append(np.array([1,0,1]))
N.append(np.array([1,0,0]))
All= P+N
converged=False
while converged == False:
    X = random.choice(All)
    inP=0
    for i in P:
        if np.array_equal(i,X):
            inP=1
            break;
    if inP==1 and perceptron(w,X)==0:
        w = w + X.reshape(3,1)
    if inP==0 and perceptron(w,X)==1:
        w = w - X.reshape(3,1)
    converged = check_convergence(w,P,N)
print("Full Convergence achieved")
```

```python
from sklearn import datasets
from sklearn import preprocessing
import time

wine = datasets.load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target
mat=df.to_numpy()
#print(wine.DESCR)
print("Shape of matrix loaded:",mat.shape)
print("Number of Features:",mat.shape[1]-1)
print("Number of Patterns:",mat.shape[0])
print(mat)
```

```python
np.random.shuffle(mat)  
X=mat[:,:-1]

X = preprocessing.normalize(X)
X=np.column_stack((np.ones(X.shape[0]),X))
Y=mat[:,-1]
Y=Y.reshape(mat.shape[0],1)
X=np.column_stack((X,Y))

train_size=int(X.shape[0]*0.4)
X_train,Y_train=X[0:train_size,:],Y[0:train_size,:]
X_test,Y_test=X[train_size:,:],Y[train_size:,:]

print(X)
```

```python
def sigmoid(X,w):
    x=X.dot(w)
    res = 1/(1+np.exp(-x))
    if res>=0.5:
        return 1
    else:
        return 0
def convergence_status(X,w):
    total = X.shape[0]
    correct=0
    for row in X:
        if sigmoid(row[:-1],w)!=row[-1]:
            correct+=1
    print("Acc ",correct*100.0/total)
    if correct==total:
        return True
    else:
        return False
def train_perceptron(x,w):
    num_samples=x.shape[0]
    converged = False
    while converged==False:
        row = x[random.randint(0,num_samples-1)]
        pred=sigmoid(row[:-1],w)
        if row[-1]==1 and pred==0:
            w+=row[:-1].reshape(row.shape[0]-1,1)
        if row[-1]==0 and pred==1:
            w-=row[:-1].reshape(row.shape[0]-1,1)
        print(w)
        converged=convergence_status(x,w)
    return w
```

```python
num_classes=3
w=[np.zeros(X.shape[1]-1).reshape(X.shape[1]-1,1) for i in range(num_classes)]
print(len(w))
for i in range(num_classes):
    x=X_train.copy()
    for row in x:
        if row[-1]==i:
            row[-1]=1
        else:
            row[-1]=0
    w[i]=train_perceptron(x,w[i])
    print("Converged ",i)
```
